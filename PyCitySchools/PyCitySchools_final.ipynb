# PyCity Schools Analysis

- Given below is an analysis of the student grades in Reading and Math.
- This data used in this analysis corresponds to the grades of all the students that attend over 15 different schools.
  
---

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = len(school_data_complete["school_name"].unique())
school_count
```

```python
# Calculate the total number of students
student_count = len(school_data_complete["student_name"])

#Formatting the values
student_total = f"{student_count:,}"
print(student_total)

#Printing original answer before formatting
student_count
```

```python
# Calculate the total budget
total_budget = school_data_complete["budget"].sum()

#Formatting the values
total_budget_formatted = f"${total_budget:,}"
print(total_budget_formatted)

#Printing original answer before formatting
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete["math_score"].mean()

#Formatting the values
average_score_math_formatted = f"{average_math_score:.2f}"
print(average_score_math_formatted)

#Printing original answer before formatting
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete["reading_score"].mean()

#Formatting the values
average_reading_score_formatted = f"{average_reading_score:.2f}"
print(average_reading_score_formatted)

#Printing original answer before formatting
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / np.int64(student_count) * 100

#Formatting the values
passing_math_percentage_formatted = f"{passing_math_percentage:.2f}%"
print(passing_math_percentage_formatted)

#Printing original answer before formatting
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / np.int64(student_count) * 100

#Formatting the values
passing_reading_percentage_formatted = f"{average_reading_score:.2f}%"
print(passing_reading_percentage_formatted)

#Printing original answer before formatting
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100

#Formatting the values
overall_passing_rate_formatted = f"{overall_passing_rate:.2f}%"
print(overall_passing_rate_formatted)

#Printing original answer before formatting
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({"Total Schools" : school_count, "Total Number of Students" : student_total,
                                 "Total Budget" : total_budget_formatted, 
                                 "Average Score (Maths)" : average_score_math_formatted,
                                 "Average Score (Reading)" : average_reading_score_formatted,
                                 "% Passing (Maths)" : passing_math_percentage_formatted,
                                 "% Passing(Reading)" : passing_reading_percentage_formatted,
                                 "Overall Passing %" : overall_passing_rate_formatted}, 
                                index=[0])

# Formatting
# district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
# district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
#For reference
school_data_complete.head()
```

```python
#Setting the School name as the index for the School Summary table
grouped_schools = school_data_complete.groupby(["school_name"])
```

```python
# Use the code provided to select all of the school types
school_types = grouped_schools["type"].first()
```

```python
# Calculate the total student count per school
per_school_counts = school_data_complete["school_name"].value_counts()
```

```python
# Calculating the total school budget per school
per_school_budget = grouped_schools["budget"].first()
```

```python
#Calculating the per capita spending per school
per_school_capita = per_school_budget/per_school_counts
```

```python
# Calculate the average test scores (Math) per school
per_school_math = grouped_schools["math_score"].mean()
```

```python
# Calculate the average test scores (Reading) per school
per_school_reading = grouped_schools["reading_score"].mean()
```

```python
# Calculate the number of students per school with math scores of 70 or higher
students_passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)]
school_students_passing_math = students_passing_math.groupby("school_name")["student_name"].count()
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
students_passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)]
school_students_passing_reading = students_passing_math.groupby("school_name")["student_name"].count()
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_students_passing_math / per_school_counts * 100
per_school_passing_reading = school_students_passing_reading / per_school_counts * 100
overall_passing_rate = school_students_passing_math_and_reading / per_school_counts * 100
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({"School Type" : school_types, "Total Number of Students" : per_school_counts, 
                                   "Total Budget" : per_school_budget, 
                                   "Per Capita Spending" : per_school_capita, 
                                   "Average Score (Maths)" : per_school_math,
                                   "Average Score (Reading)" : per_school_reading, 
                                   "% Passing (Math)" : per_school_passing_math,
                                   "% Passing (Reading)" : per_school_passing_reading,
                                   "Overall Passing %" : overall_passing_rate})

# Formatting
per_school_summary["Total Number of Students"] = per_school_summary["Total Number of Students"].map("{:,}".format)
per_school_summary["Total Budget"] = per_school_summary["Total Budget"].map("${:,.2f}".format)
per_school_summary["Per Capita Spending"] = per_school_summary["Per Capita Spending"].map("${:,.2f}".format)
per_school_summary["Average Score (Maths)"] = per_school_summary["Average Score (Maths)"].map("{:,.2f}".format)
per_school_summary["Average Score (Reading)"] = per_school_summary["Average Score (Reading)"].map("{:,.2f}".format)
per_school_summary["% Passing (Math)"] = per_school_summary["% Passing (Math)"].map("{:.2f}%".format)
per_school_summary["% Passing (Reading)"] = per_school_summary["% Passing (Reading)"].map("{:.2f}%".format)
per_school_summary["Overall Passing %"] = per_school_summary["Overall Passing %"].map("{:.2f}%".format)

# Display the DataFrame
per_school_summary
```

## Reading Score by Grade 

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the the `reading_score` column for each.
ninth_grade_reading_scores = ninth_graders.groupby("school_name")["reading_score"].mean().map("{:,.4f}".format)
tenth_grader_reading_scores = tenth_graders.groupby("school_name")["reading_score"].mean().map("{:,.4f}".format)
eleventh_grader_reading_scores = eleventh_graders.groupby("school_name")["reading_score"].mean().map("{:,.4f}".format)
twelfth_grader_reading_scores = twelfth_graders.groupby("school_name")["reading_score"].mean().map("{:,.4f}".format)

# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade = pd.DataFrame({"9th": ninth_grade_reading_scores, "10th": tenth_grader_reading_scores,
                                        "11th": eleventh_grader_reading_scores, "12th": twelfth_grader_reading_scores})

# Minor data wrangling
reading_scores_by_grade = reading_scores_by_grade[["9th", "10th", "11th", "12th"]]
reading_scores_by_grade.index.name = "Reading Scores per Grade"

# Display the DataFrame
reading_scores_by_grade
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_budget/per_school_counts,
                                                             spending_bins, labels=labels)
school_spending_df
```

```python
school_spending_df["Average Score (Maths)"] = school_spending_df["Average Score (Maths)"].apply(lambda x: x.replace(',', '')).astype('float')
school_spending_df["Average Score (Reading)"] = school_spending_df["Average Score (Reading)"].apply(lambda x: x.replace(',', '')).astype('float')
school_spending_df["% Passing (Math)"] = school_spending_df["% Passing (Math)"].apply(lambda x: x.replace('%', '').replace(',', '')).astype('float')
school_spending_df["% Passing (Reading)"] = school_spending_df["% Passing (Reading)"].apply(lambda x: x.replace('%', '').replace(',', '')).astype('float')
school_spending_df["Overall Passing %"] = school_spending_df["Overall Passing %"].apply(lambda x: x.replace('%', '').replace(',', '')).astype('float')
```

```python
#  Calculate averages for the desired columns. 
grouped_data_df = school_spending_df.groupby(["Spending Ranges (Per Student)"])

spending_math_scores = grouped_data_df["Average Score (Maths)"].mean()

spending_reading_scores = grouped_data_df["Average Score (Reading)"].mean()

spending_passing_math = grouped_data_df["% Passing (Math)"].mean()

spending_passing_reading = grouped_data_df["% Passing (Reading)"].mean()

overall_passing_spending = grouped_data_df["Overall Passing %"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({"Average Score (Maths)":spending_math_scores,
                                 "Average Score (Reading)":spending_reading_scores,
                                 "% Passing (Math)":spending_passing_math,
                                 "% Passing (Reading)":spending_passing_reading, 
                                 "Overall Passing %":overall_passing_spending})

spending_summary.index.name = "Scores by School Spending"

# Display results
spending_summary
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(school_data_complete["school_name"].value_counts(),
                                           size_bins, labels=labels)
                                           
per_school_summary
```

```python
per_school_summary["Average Score (Maths)"] = per_school_summary["Average Score (Maths)"].apply(lambda x: x.replace(',', '')).astype('float')
per_school_summary["Average Score (Reading)"] = per_school_summary["Average Score (Reading)"].apply(lambda x: x.replace(',', '')).astype('float')
per_school_summary["% Passing (Math)"] = per_school_summary["% Passing (Math)"].apply(lambda x: x.replace('%', '').replace(',', '')).astype('float')
per_school_summary["% Passing (Reading)"] = per_school_summary["% Passing (Reading)"].apply(lambda x: x.replace('%', '').replace(',', '')).astype('float')
per_school_summary["Overall Passing %"] = per_school_summary["Overall Passing %"].apply(lambda x: x.replace('%', '').replace(',', '')).astype('float')
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])["Average Score (Maths)"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Score (Reading)"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing (Math)"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing (Reading)"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["Overall Passing %"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({"Average Score (Maths)":size_math_scores,
                             "Average Score (Reading)":size_reading_scores,
                             "% Passing (Math)":size_passing_math,
                             "% Passing (Reading)":size_passing_reading,
                             "Overall Passing %":size_overall_passing})

size_summary.index.name = "Scores by School Size"

# Display results
size_summary
```

## Scores by School Type

```python
# Group the per_school_summary DataFrame by "School Type" and average the results.
average_math_score_by_type = per_school_summary.groupby(["School Type"])["Average Score (Maths)"].mean()
average_reading_score_by_type = per_school_summary.groupby(["School Type"])["Average Score (Reading)"].mean()
average_percent_passing_math_by_type = per_school_summary.groupby(["School Type"])["% Passing (Math)"].mean()
average_percent_passing_reading_by_type = per_school_summary.groupby(["School Type"])["% Passing (Reading)"].mean()
average_percent_overall_passing_by_type = per_school_summary.groupby(["School Type"])["Overall Passing %"].mean()
```

```python
# Assemble the new data by type into a DataFrame called `type_summary`
type_summary = pd.DataFrame({"Average Score (Maths)":average_math_score_by_type,
                             "Average Score (Reading)":average_reading_score_by_type,
                             "% Passing (Math)":average_percent_passing_math_by_type,
                             "% Passing (Reading)":average_percent_passing_reading_by_type,
                             "Overall Passing %":average_percent_overall_passing_by_type})

type_summary.index.name = "Scores by School Type"

# Display results
type_summary
```
